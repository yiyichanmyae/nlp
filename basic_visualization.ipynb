```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. simple plot with numbers list

A single list works in .plot() without x and y lists. It plots the linechart which puts the given values as y values.

```python
plt.plot([1,5,6,8,11])
plt.show()
```

### 2. Points have x and y values. Add titles and axis labels

.plot() also works by giving (x list,y list). it's clear as x  is x and y is y.

```python
plt.plot([1,2,3,4], [4,2,9, 16])
plt.title('X and Y test plot')
plt.xlabel('x values')
plt.ylabel('y valeus')
plt.show()
```

### 3. Show points with red dots. Define Figure size. Set axis scales. Put annotation.

normally .plot() shows line chart. but you can define the styles to show with the formatted string.


    fmt = '[marker][line][color]'
    
example;    
'ro'      # red o

'bx'      # black(color) x(marker)

'--co'     # --(line) c(color) o(marker)  

```python
plt.figure(figsize=(2,5))              #  x wides 2 inches, y with 5 inches tall
plt.plot([1,2,3,4], [4,2,9, 16], '--oc') # cyan o
plt.axis([0, 5, 0, 18])                # xmin, xmax, ymin, ymax
plt.annotate('anno', (3,6))
plt.show()
```

### 4. bar chart with colour bars

well, .bar() does not work with a single list, like .plot() does. so we need x and y lists.

```python
plt.clf()                         # clear figure
x = [7,4,5,6]
y = [1,4,6,3]
plt.xticks(x, ['Anka', 'Anders', 'Andrey', 'Andrew'])
# plt.bar(x,y)
# plt.bar(x, y, color='y' )
plt.bar(x, y, color=['r', 'y', 'c', 'pink'])
plt.show()
```

### 5. ploting 2 columns of Dataframe/ 2 different list (similar with lineplot)

```python
dt = {
    'Red o' : np.random.rand(10),
    'green x' : np.random.rand(10)
}

dt_df = pd.DataFrame(dt)

dt_df.plot(style=['ro', 'gx'])
# plt.show()
```

### 6. plotting TimeSeries data

```python
df = pd.DataFrame(np.random.randn(180, 3), index=pd.date_range('1/1/2018', periods=180), columns=list('NGE'))
df.cumsum().plot()
```

### 7. scatter plot with colouring, sizes and transparency 

```python
N = 80

x = np.random.randn(N)
y = np.random.randn(N)
sizes = (N/5*np.random.randn(N))**2
colors = np.random.randn(N)

plt.scatter(x, y, sizes=sizes, c=colors, alpha=0.5)
plt.show()
```
